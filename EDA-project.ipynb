### Data Preprocessing/Data Transformation:

```python
# Filtering student's parental level of education segment:
filtered_student = dq[(dq['parental_level_of_education'] == "bachelor's degree")]
print(filtered_student)
```

```python
# Filtering not a consumer segment:
filtered_not = dq[~(dq['parental_level_of_education'] == "bachelor's degree")]
print(filtered_not)
```

```python
filtered = dq[(dq['math_score'] > 50) & (dq['reading_score'] >50) & (dq['writing_score'] >50)]
print(filtered)
```

```python
filtered = dq[(dq['math_score'] > 50) | (dq['reading_score'] >50) | (dq['writing_score'] >50)]
print(filtered)
```

```python
# sorting:
sorted_data = dq.sort_values(by='race_ethnicity', ascending=False)
print(sorted_data)
```

#### Histogram

```python
dq.describe()
```

```python
bins=np.arange(dq['math_score'].min(),dq['math_score'].max()+30,30)
counts,bins,patches=plt.hist(dq['math_score'],bins=bins)
print(counts) #[91. 41. 32. 28. 12.  3.  1.]
print(bins) #[  0.  30.  60.  90. 120. 150. 180. 210.]
print(bins[1]) # index of the bins
bin_width=bins[1]-bins[0]
print(len(counts))

for i in range(len(counts)): # len(counts) = 7
    plt.text(bins[i]+bin_width/2,counts[i],str(int(counts[i])),ha='center',va='bottom',fontsize=12)
    plt.title("Histogram on Maths Score")
plt.xlabel("Math score") # X-axis how much is the score from 0 to last run
plt.ylabel("Frequency") # Y-axis how frequently the person scored
plt.xticks(bins,rotation=45) # X-axis value edits - eg: no '30' in x-axis is rotated in 45 degree
plt.show()
```

```python
Q1=dq['math_score'].quantile(0.25)
Q3=dq['math_score'].quantile(0.75)
minimum=dq['math_score'].min()
plt.boxplot(dq['math_score'])
plt.text(x=1.1,y = dq['math_score'].min(),s = f"min: {minimum}")
plt.text(x=1.1,y = dq.math_score.quantile(0.25),s = f"Q1 -> {Q1}")
plt.text(x=1.1,y = dq['math_score'].median(),s = f"median: (Q2)")
plt.text(x=1.1,y = dq.math_score.quantile(0.75),s = f"Q3 -> {Q3}")
plt.text(x=1.1,y = dq['math_score'].max(),s = f"max")
#the graph tittle and axes labels
plt.title('Boxplot of Math Score')
plt.ylabel('math score')
plt.show()
```

```python
# Sorted countplot:
plt.figure(figsize=(8,6))
sns.countplot(data=df,y="parental_level_of_education",order=df['parental_level_of_education'].value_counts().index,palette="pastel")
plt.title("Frequency of parental level of education ",fontsize=16)
plt.xlabel("count",fontsize=12)
plt.ylabel("parental level of education Mode",fontsize=12)
plt.show()
```

```python
# Customization:
sns.set(style="whitegrid")
# Scatterplot: Runs vs. Balls Faced
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dq, x="reading_score", y="writing_score", hue="gender" ,palette="viridis")
# other palette -'plasma', 'coolwarm', 'cividis',"viridis"
plt.title("writing score vs. reading score	", fontsize=16)
plt.xlabel("reading score	", fontsize=12)
plt.ylabel("writing score", fontsize=12)
plt.legend(title="gender", loc='upper left')
plt.show()
```
